```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("hotel_booking.csv")
```

```python
df.shape
```

```python
print(df.head())
```

```python
df.isnull().sum()
```

```python
df = df.drop(['agent','company'], axis=1)
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns = {'stays_in_weekend_nights' : 'num_of_weekend_nights', 'stays_in_week_nights' : 'num_of_week_nights', 'adr' : 'average_daily_rate'})
```

```python
print(df.head(5))
```

```python
df['arrival_date_year'] = df['arrival_date_year'].astype(str)
df['arrival_date_month'] = df['arrival_date_month'].astype(str)
df['arrival_date_day_of_month'] = df['arrival_date_day_of_month'].astype(str)

```

```python
df['arrival_date'] = df['arrival_date_year'] + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month']
```

```python
df['arrival_date'] = pd.to_datetime(df['arrival_date'])
```

```python
df.drop(['arrival_date_year','arrival_date_month','arrival_date_day_of_month'],axis=1)
```

```python
df = df[['name','email','phone-number','credit_card','arrival_date','reservation_status_date','reservation_status'] + list(df.columns[:-8])]
```

```python
print(df.head(5))
```

```python
df['is_canceled'] = df['is_canceled'].astype(bool)
df['is_repeated_guest'] = df['is_repeated_guest'].astype(bool)
```

```python
print(df['is_canceled'])
```

```python
duplicates = df[df.duplicated()]

print(duplicates)
```

```python
df['children'] = df['children'].astype(int)
```

```python
df['average_daily_rate'] = '$' + df['average_daily_rate'].astype(float).round(2).astype(str)
```

```python
df.drop(['arrival_date_week_number'],axis=1)
```

```python
df.to_csv('cleaned_data.csv')
```
